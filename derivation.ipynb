```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
  return 4*x+np.cos(x)-2
```

```python
def de(x):
  return 4-np.sin(x)
```

```python
true_value = de(3)
```

```python
def fdd(f,x,h):
  return (f(x+h)-f(x))/h
```

```python
h=[0.3,0.1,0.15,0.001]
x=3
fdd_errors = []
for i in h:
  approx_value=fdd(f,x,i)
  arte=abs(true_value-approx_value)
  fdd_errors.append(arte)
```

```python
print(fdd_errors)
```

```python
def bdd(f,x,h):
  return(f(x)-f(x-h))/h

```

```python
h=[0.3,0.1,0.15,0.001]
x=3
bdd_errors = []
for i in h:
  approx_value=bdd(f,x,i)
  arte=abs(true_value-approx_value)
  bdd_errors.append(arte)
```

```python
print(bdd_errors)
```

```python
def cdd(f,x,h):
  return (f(x+h)-f(x-h))/2*h
```

```python
h=[0.3,0.1,0.15,0.001]
x=3
cdd_errors = []
for i in h:
  print(i)
  approx_value=cdd(f,x,i)
  arte=abs(true_value-approx_value)
  cdd_errors.append(arte)
```

```python
print(cdd_errors)
```

```python
plt.plot(h,fdd_errors,label='FDD')
plt.plot(h,bdd_errors,label='BDD')
plt.plot(h,cdd_errors,label='CDD')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(h,fdd_errors,label='FDD',marker='o')
plt.plot(h,bdd_errors,label='BDD',marker='^')
plt.plot(h,cdd_errors,label='CDD',marker='x')
plt.xlabel('h',fontsize=14)
plt.ylabel('error',fontsize=14)
plt.legend()
plt.show()
```

```python
x=np.linspace(-1,1,100)
h=0.01
der_fdd=fdd(f,x,h)
der_bdd=bdd(f,x,h)
der_cdd=cdd(f,x,h)
der_true=de(x)
plt.figure(figsize=(10,5))
plt.plot(x,der_fdd,label='FDD')
plt.plot(x,der_bdd,label='BDD')
plt.plot(x,der_cdd,label='CDD')
plt.plot(x,der_true,label='True')
plt.legend()
plt.show()
```
