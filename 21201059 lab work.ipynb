```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def f(x):
  return 4*(x**5)-3*(x**3)+2*x-9
```

```python
def de(x):
  return 20*(x**4)-9*(x**2)+2
```

```python
previous_value = 15.3627245864055853
```

```python
true_value = de(3)
```

```python
def fdd(f,x,h):
  return (f(x+h)-f(x))/h
```

```python
h=[0.3,0.15,0.1,0.05,0.01,0.001]
x=3
fdd_errors = []
for i in h:
  approx_value=fdd(f,x,i)
  arte =  abs( true_value-approx_value)
  fdd_errors.append(arte)
print(fdd_errors)
```

```python
def bdd(f,x,h):
  return (f(x)-f(x-h))/h
```

```python
h=[0.3,0.15,0.1,0.05,0.01,0.001]
x=3
bdd_errors = []
for i in h:
  approx_value = bdd(f,x,i)
  arte = abs( true_value-approx_value)
  bdd_errors.append(arte)
print(bdd_errors)
```

```python
def cdd(f,x,h):
  return (f(x+h)-f(x-h))/(2*h)
```

```python
h=[0.3,0.15,0.1,0.05,0.01,0.001]
x=3
cdd_errors=[]
for i in h:
  print(i)
  approx_value = cdd(f,x,i)
  arte =abs( true_value-approx_value)
  cdd_errors.append(arte)
  #print("h = ",i,"; derivative=", approx_value,";ARTE=,",arte,"%")
  print(cdd_errors)
```

```python
plt.plot(h,fdd_errors,label='FDD')
plt.plot(h,bdd_errors,label='BDD')
plt.plot(h,cdd_errors,label='CDD')
plt.xlabel('h')
plt.ylabel('Approximation_error')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(h,fdd_errors,label='FDD',marker='o')
plt.plot(h,bdd_errors,label='BDD',marker='^')
plt.plot(h,cdd_errors,label='CDD',marker='x')
plt.xlabel('h',fontsize=14)
plt.ylabel('$|\epsilon_t|$%',fontsize=14)
plt.legend()
plt.show()
```

```python
x =np.linspace(-1,1,1000)
h=0.01
der_fdd = fdd(f,x,h)
der_bdd = bdd(f,x,h)
der_cdd = cdd(f,x,h)
der_true = de(x)

plt.figure(figsize=(10,5))
plt.plot(x,der_fdd,label='FDD')
plt.plot(x,der_bdd,label='BDD')
plt.plot(x,der_cdd,label='CDD')
plt.plot(x,der_true,label='True')



plt.xlabel('$x$',fontsize=15)
plt.ylabel('$f\'(x)$',fontsize=15)
plt.show()
```
