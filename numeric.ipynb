```python
#harmonic function

def har(n):
  a=[]
  for i in range(1,n+1):
    #a.append(1/i)
    term=1/i
    a.append(term)
  return a
```

```python
print(har(4))
```

```python
sum_first_n_terms=np.sum(har(4))
sum_first_n_terms
```

```python
def har_mean(n):
  first_n_terms = har(n)
  sum_of_n_terms = np.sum(first_n_terms)
  mean=n/sum_of_n_terms
  return mean
```

```python
print(har_mean(4))
```
